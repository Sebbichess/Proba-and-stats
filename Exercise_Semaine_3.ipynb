## 1. The wearing out of some gears was measured in 10 -4 mm 3, when using different lubrication oils having different viscosities in mPa-s. The data produced the following table:

```python
'''
Import here useful libraries
Run this cell first for convenience
'''
import numpy as np
from scipy import stats
import scipy
import warnings
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from scipy.ndimage import mean, median, variance
warnings.simplefilter('ignore', DeprecationWarning)
```

```python
data1 = np.array([58.8,30.8,27.3,29.9,17.7,76.5])
data1 = data1.reshape(len(data1), 1) 
df1 = pd.DataFrame(data=data1, columns=['20.4mPa-s']) # Transform data to pandas DataFrame
print('Data1: \n', df1)
data2 = np.array([44.5,47.1,48.7,41.6,32.8,18.3])
data2 = data2.reshape(len(data2), 1) 
df2 = pd.DataFrame(data=data1, columns=['30.2mPa-s']) # Transform data to pandas DataFrame
print('Data2: \n', df2)
data3 = np.array([73.3,57.1,66,93.8,133.2,81.1])
data3 = data3.reshape(len(data3), 1) 
df3 = pd.DataFrame(data=data3, columns=['89.4mPa-s']) # Transform data to pandas DataFrame
print('Data3: \n', df3)
data4 = np.array([30.6,24.2,16.6,38.9,28.7,23.6])
data1 = data4.reshape(len(data4), 1) 
df4 = pd.DataFrame(data=data4, columns=['252.6mPa-s']) # Transform data to pandas DataFrame
print('Data4: \n', df4)
```

```python
Q11 = df1.quantile(0.25)
Q31 = df1.quantile(0.75)
IQR1 = Q31 - Q11
#print(IQR)
# We print the outliers here
mask = ((df1 < (Q11 - 1.5 * IQR1)) |(df1 > (Q31 + 1.5 * IQR1)))
filtered_data1 = (mask*df1).to_numpy()
print("Outliers:")
for datum in filtered_data1:
    if(datum != 0):
        print(datum)
Q12 = df2.quantile(0.25)
Q32 = df2.quantile(0.75)
IQR2 = Q32 - Q12
#print(IQR)
# We print the outliers here
mask = ((df2 < (Q12 - 1.5 * IQR2)) |(df2 > (Q32 + 1.5 * IQR2)))
filtered_data2 = (mask*df2).to_numpy()
print("Outliers:")
for datum in filtered_data2:
    if(datum != 0):
        print(datum)
Q13 = df3.quantile(0.25)
Q33 = df3.quantile(0.75)
IQR3 = Q33 - Q13
#print(IQR)
# We print the outliers here
mask = ((df3 < (Q13 - 1.5 * IQR3)) |(df3 > (Q33 + 1.5 * IQR3)))
filtered_data3 = (mask*df3).to_numpy()
print("Outliers:")
for datum in filtered_data3:
    if(datum != 0):
        print(datum)
Q14 = df4.quantile(0.25)
Q34 = df4.quantile(0.75)
IQR4 = Q34 - Q14
#print(IQR)
# We print the outliers here
mask = ((df4 < (Q14 - 1.5 * IQR4)) |(df4 > (Q34 + 1.5 * IQR4)))
filtered_data4 = (mask*df4).to_numpy()
print("Outliers:")
for datum in filtered_data4:
    if(datum != 0):
        print(datum)
```

```python
print("Sample mean i: ", mean(df1))
print("Sample median i: ", median(df1))
print("Sample mean ii: ", mean(df2))
print("Sample median ii: ", median(df2))
print("Sample mean iii: ", mean(df3))
print("Sample median iii: ", median(df3))
print("Sample mean iv: ", mean(df4))
print("Sample median iv: ", median(df4))
```
