1.	Exposure to microbial products, especially endotoxin, may have an impact on vulnerability to allergic diseases.  The following data on concentration (EU/mg) in settled dust for one sample of urban homes and another of farm homes was obtained: 

U: 6.0 5.0 11.0 33.0 4.0 5.0 80.0 18.0 35.0 17.0 23.0

F: 4.0 14.0 11.0 9.0 9.0 8.0 4.0 20.0 5.0 8.9 21.0 9.2 3.0 2.0 0.3


(a) Determine the sample mean for each sample. How do they compare?

```python
# The following code defines two arrays.for U and F respectively.    
# Add in the relevant code to print the mean

U = np.array([6.0, 5.0, 11.0, 33.0, 4.0, 5.0, 80.0, 18.0, 35.0, 17.0, 23.0])
F = np.array([4.0, 14.0, 11.0, 9.0, 9.0, 8.0, 4.0, 20.0, 5.0, 8.9, 21.0, 9.2, 3.0, 2.0, 0.3])
#####  Part a code below
print("U mean: ", mean(U))
print("F mean: ", mean(F))
#####
```

(b) Determine the sample median for each sample. How do they compare? Why is the median for the urban sample so different from the mean for that sample?

```python
######## part b code below
print("U median: ", median(U))
print("F median: ", median(F))
#######
```

(c) Calculate the trimmed mean for each sample by deleting the smallest and largest observation. What are the corresponding trimming percentages? How do the values of these trimmed means compare to the corresponding means and medians?

```python
#### code for part c
r = 0.1 # r value for trimming
print("U trimmed mean: ", stats.trim_mean(U, r))
print("F trimmed mean: ", stats.trim_mean(F, r))

#####
```
