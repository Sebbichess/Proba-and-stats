## Considérerations énergétiques $E_{cin} =\frac{1}{2} m v^2$


```python
Vmaxkph = 15
Vmaxmps = Vmaxkph / 3.6
coeff_marg = 2
masse = 5 
Epot_necessaire = 1/2 * masse * (Vmaxmps * coeff_marg)**2
print(f'Energie cinetique necessaire : {Epot_necessaire} J avec un coefficient de marge de {coeff_marg} sur la vitesse max')
```

1. Ressort simple $E_{pot} = \frac{1}{2} k \Delta x^2$

On fait la simplification que la longeur max du ressort est 2x sa longeur initiale et on a un distance de 400mm disponible calculons la la constante k de ressort necessaire

```python
d_dispo = 0.4
k = 2 * Epot_necessaire /(d_dispo/2)**2
print(f'Il nous faut des ressorts en parrallele ayant au moins une constante de rappel k de {k} N/m pour que ce système de stockage soit viable ')
```

2. Ressort Force constante $E_{pot} = F * d $

```python
F = Epot_necessaire/d_dispo
print(f'Il nous faut des ressorts a Force constante qui ont une force totale de au moins {F} N pour que ce système de stockage soit viable ')
```

Ce resultat semble peut être demande une grande force, mais nous avons une solutions a ce problème

3. Cylindre d'inertie (flywheel) $E_{pot} = \frac{1}{2} I \omega ^2 $

$I =\frac{1}{2} m R^2$

```python
#on suppose un masse du cylindre de 4 kg et un rayon de 100mm (le plus grand possible)
Imax = 4/2 * (0.1)**2
omega = math.sqrt(2 * Epot_necessaire/ Imax)
tm = omega * 60 /(2 *math.pi)
print(f'Il faudrait que le cylindre atteingnent un vitesse angulaire de w = {omega} rad/s ou soit {tm} tours/minutes')
```

Etant une vitesse angulaire élevée, les frottements mecaniques le sont tout autant, ils nous saura donc trop compliqué de mettre cette idée en place, nous allons donc l'éliminée

4. Chute (bloc qui tombe) $ E_{pot} = mgh$

```python
#On pose nos dimensions verticalement en pour maximiser la hauteur de chute du bloc  
h = 0.4
m = Epot_necessaire /(9.81 * h)

print(f'Il nous faudrait une masse de bloc de au moins {m} kg')
```
